```python
import pandas as pd
import numpy as np
```

```python
# 데이터 로드
bmi_data_path = "data/bmi_data.csv" 
food_data_path = "data/food_data.csv"
exercise_data_path = "data/exercise_data.csv"
```

```python
bmi_data = pd.read_csv(bmi_data_path)
food_data = pd.read_csv(food_data_path)
exercise_data = pd.read_csv(exercise_data_path)
```

```python
# 📌 모든 user_id 리스트 (bmi_data 기준)
all_users = set(bmi_data["user_id"])
```

```python
# 📌 사용자가 섭취한 음식 확인
food_data["섭취 여부"] = food_data["user_id"].apply(lambda x: "1" if x in all_users else "0")
```

```python
food_data.to_csv("./data/food_data.csv", index=False)
```
